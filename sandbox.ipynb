```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
```

```python
import google.generativeai as genai
for genmodel in genai.list_models():
    if "embedding" in genmodel.name:
        print(genmodel.name)
```

```python
api_key = os.getenv("GOOGLE_API_KEY")
```

```python
llm = ChatGoogleGenerativeAI(model="gemma-4-26b-a4b-it", google_api_key=api_key)
```

```python
prompt = ChatPromptTemplate.from_template("Answer this question: {question}")

# Chain prompt + model together
chain = prompt | llm

# Run it!
result = chain.invoke({"question": "What is the capital of France?"})
print(result.content)
```

```python
for k, v in result.model_dump().items():
    print(f"{k}: {v}")
```

```python

import os
load_dotenv()
path = os.getenv("PATH_TO_PROJECT_FOLDER")

extensions = [".py", ".js", ".ts", ".java", ".cpp", ".go", ".rs"]

for root, dirs, files in os.walk(path):
    for file in files:
        for ext in extensions:
            if file.endswith(ext):
                filepath = os.path.join(root, file)
                try:
                    with open(filepath, "r", encoding="utf-8") as f:
                        f.read()
                except UnicodeDecodeError as e:
                    print(f"Problem file: {filepath} ({ext})")
```
